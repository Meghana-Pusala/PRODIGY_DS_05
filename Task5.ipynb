# Creating visualizations

```python
#Distribution of Accident severity
traffic_data['Accident_severity'].value_counts()
```

```python
#plotting the final class
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x = traffic_data['Accident_severity'])
plt.title('Distribution of Accident severity')
```

Analysis:

Slight Injury: Most common type of accident, with around 10,000 cases.

Serious Injury: Much less common, with around 2,000-3,000 cases.

Fatal Injury: Rarest type of accident, with very few cases (close to 0).

```python
plt.figure(figsize=(8, 4))
sns.countplot(y='Weather_conditions', data=traffic_data)
plt.title('Accidents by Weather Conditions')
plt.xlabel('Number of Accidents')
plt.ylabel('Weather Conditions')
plt.grid(True)
plt.show()
```

Analysis:

Normal Weather: Most accidents occur in normal weather conditions, with around 10,000 cases.

Raining: A significant number of accidents happen when it’s raining, with approximately 2,000 cases.

Other Conditions (Raining and Windy, Cloudy, Windy, Snow, Fog or Mist): These conditions see very few accidents, with each category having much fewer cases than raining or normal conditions.

Unknown: There are some accidents where the weather condition is unknown, with around 500-1000 cases.

```python
plt.figure(figsize=(8, 4))
sns.countplot(y='Road_surface_conditions', data=traffic_data, hue='Road_surface_conditions', palette='viridis', dodge=False, legend=False)
plt.title('Accidents by Road Surface Conditions')
plt.xlabel('Number of Accidents')
plt.ylabel('Road Surface Conditions')
plt.grid(True)
plt.show()
```

This bar chart illustrates the number of accidents that occurred under different road surface conditions. The conditions are listed on the vertical axis, while the number of accidents is shown on the horizontal axis.

Dry conditions had the highest number of accidents, with well over 8,000 incidents.

Wet or damp conditions followed, with a significant drop, showing around 4,000 accidents.

Snow saw a much lower number of accidents, with only a small number, close to 100.

Flood over 3 cm deep had no reported accidents in the data.

```python
plt.figure(figsize=(8, 6))
corr = traffic_data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Traffic Accident Data')
plt.show()
```

-This heatmap shows the correlation between two variables in the traffic accident data: the Number of vehicles involved and the Number of casualties.

*The diagonal values (1.00) show that each variable is perfectly correlated with itself, which is expected.

*The off-diagonal values (0.21) represent the correlation between the number of vehicles involved in an accident and the number of casualties. A correlation value of 0.21 indicates a weak positive correlation between these two variables.

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='Type_of_vehicle', data=traffic_data,hue='Type_of_vehicle', palette='pastel',dodge=False, legend=False)
plt.title('Accidents by Vehicle Type')
plt.xlabel('Number of Accidents')
plt.ylabel('Type of Vehicle')
plt.grid(True)
plt.show()
```

*The above bar chart displays the frequency of various types of vehicles involved in accidents.

-Automobiles are the most frequently involved, with over 3,000 incidents.

-Station wagons and Lorries (41?100Q) also have a significant number of accidents, with numbers close to or exceeding 1,500.

-Public vehicles (12 seats) and Public vehicles (> 45 seats) also have a notable presence, indicating they are involved in accidents fairly often.

-Other categories like Long lorry, Taxi, and Pick up up to 10Q are involved less frequently, but still have a considerable number of accidents.

-Less common vehicles like Motorcycles, Special vehicles, Ridden horses, Bicycles, and Bajaj have the fewest accidents, with numbers well below 500.

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Light_conditions',hue='Light_conditions',data=traffic_data, palette='muted',dodge=False, legend=False)
plt.title('Accidents by Light Conditions')
plt.xlabel('Light Conditions')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()
```

Analysis:

Daylight Accidents: The highest number of accidents occur during daylight, with the bar representing over 8,000 accidents.

Darkness with Lit Lights: The number of accidents drops significantly when it's dark, but lights are lit, with fewer than 4,000 accidents.

Darkness with No Lighting: Accidents occur much less frequently in darkness with no lighting compared to daylight or lit conditions.

Darkness with Unlit Lights: This condition has the fewest accidents, with the bar being very small.

```python
# Select relevant numerical columns
numerical_data = traffic_data[['Time', 'Number_of_casualties', 'Number_of_vehicles_involved']]

# Pair plot
sns.pairplot(numerical_data)
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.boxplot(x='Accident_severity', y='Number_of_casualties', data=traffic_data, color=sns.color_palette('Set3')[0])
plt.title('Number of Casualties by Accident Severity')
plt.xlabel('Accident Severity')
plt.ylabel('Number of Casualties')
plt.grid(True)
plt.show()
```

Analysis:

Slight Injury: There are about 2 casualties resulting in slight injuries, which is the lowest number of casualties among the three categories.

Serious Injury: Similar to slight injuries, there are about 2 casualties resulting in serious injuries.

Fatal Injury: The number of casualties is significantly higher for fatal injuries, with about 4 casualties.

```python
sns.jointplot(x='Time', y='Number_of_vehicles_involved', data=traffic_data, kind='scatter', color='purple')
plt.suptitle('Time of Day vs. Number of Vehicles Involved', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Accident_severity', y='Age_band_of_driver', hue='Accident_severity', data=traffic_data, palette='pastel', legend=False)
plt.title('Driver Age by Accident Severity')
plt.xlabel('Accident Severity')
plt.ylabel('Age Band of Driver')
plt.grid(True)
plt.show()
```


* X-Axis: Represents the severity of accidents, categorized into "Slight Injury," "Serious Injury," and "Fatal Injury.

* Y-Axis: Shows the age bands of drivers involved, which are categorized as "Under 18," "18-30," "31-50," "Over 51," and "Unknown.

* "Shape of the Plots (Violin Plots):

-The width of the plot at any point indicates the density of data points in that age band for a specific injury severity.

-A wider plot suggests a higher concentration of drivers in that age group and injury severity.

* Observations

-Slight Injury: Most common among drivers aged 18-30, followed by those aged 31-50.

-Serious Injury: Predominantly occurs among the 18-30 and 31-50 age bands.

-Fatal Injury: Shows more spread, with higher occurrences in drivers aged 31-50 and significant instances in drivers over 51.

-Young and Older Drivers: Both younger (under 18) and older drivers (over 51) seem to have fewer occurrences overall, but older drivers have a notable presence in fatal injuries.
